REVISIÓN DE VALORES EXTREMOS (OUTLIERS):

```python
columnas_outliers = ['ABV',
       'Min IBU', 'Max IBU', 'Astringency', 'Body', 'Alcohol', 'Bitter',
       'Sweet', 'Sour', 'Salty', 'Fruits', 'Hoppy', 'Spices', 'Malty',
       'review_aroma', 'review_appearance', 'review_palate', 'review_taste',
       'review_overall']

n_filas = beer.shape[0]

for col in columnas_outliers:

 Q1 = beer[col].quantile(0.25)
 Q3 = beer[col].quantile(0.75)
 IQR = Q3 - Q1

 limite_inferior = Q1 - 1.5 * IQR
 limite_superior = Q3 + 1.5 * IQR

 outliers = beer[(beer[col] < limite_inferior) | (beer[col] > limite_superior)]


 moda = beer[col].mode()[0]
 beer[col] = beer[col].mask((beer[col] < limite_inferior) | (beer[col] > limite_superior), moda)

 n_outliers = len(outliers)
 porcentaje = (n_outliers / n_filas) * 100

 print(f"La columna {col} tiene {n_outliers} outliers, que representan el {porcentaje:.2f}% del total de filas.")

 beer.boxplot(column=columnas_outliers, rot=90, figsize=(8,4))
 
plt.title("Boxplot de las columnas numéricas")
plt.show()
outliers
```

```python
columnas_numericas = beer.select_dtypes(include=['int', 'float']).columns

# Itera sobre cada columna numérica y aplica la transformación de Box-Cox
for columna in columnas_numericas:
    # Extrae los datos de la columna actual
    datos_columna = beer[columna]
    
    # Aplica la transformación de Box-Cox a los datos de la columna
    datos_transformados_boxcox, lambda_boxcox = boxcox(datos_columna + 1)  # Agrega 1 para evitar ceros
    
    # Actualiza los datos transformados en el DataFrame
    beer[columna] = datos_transformados_boxcox

# Imprime el DataFrame con los datos transformados
beer
```

```python
variables = beer.drop(['Name',	'Style',	'Brewery',	'Beer Name (Full)',	'Description'], axis=1)
variables.columns
```

```python
# Lista de nombres de variables
variables_name = ['ABV', 'Min IBU', 'Max IBU', 'Astringency', 'Body', 'Alcohol', 'Bitter',
       'Sweet', 'Sour', 'Salty', 'Fruits', 'Hoppy', 'Spices', 'Malty',
       'review_aroma', 'review_appearance', 'review_palate', 'review_taste',
       'review_overall']

num_subplots = len(variables_name) * 3

figsize = (15, 5 * len(variables_name))  # Tamaño de la figura

fig, axs = plt.subplots(len(variables_name), 3, figsize=figsize)

# Lista de estadísticas descriptivas
stats = ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max']

# Diccionario de significados de estadísticas
stats_meaning = {
    'count': 'Número de observaciones',
    'mean': 'Media',
    'std': 'Desviación estándar',
    'min': 'Valor mínimo',
    '25%': 'Primer cuartil',
    '50%': 'Mediana',
    '75%': 'Tercer cuartil',
    'max': 'Valor máximo'
}

for i, variable in enumerate(variables_name):
    # Histograma y KDE
    sns.histplot(x=variable, data=variables, kde=True, color="green", ax=axs[i, 0])
    axs[i, 0].set_title(f'Histograma y KDE de {variable}', fontsize=14)

    # Diagrama de caja
    sns.boxplot(x=variable, data=variables, color="green", ax=axs[i, 1])
    axs[i, 1].set_title(f'Diagrama de Caja de {variable}', fontsize=14)

    # Cuadro de datos estadísticos
    des_variable = variables[variable].describe()
    df_des_variable = pd.DataFrame(des_variable)
    df_des_variable.rename(index=stats_meaning, columns={variable: 'Estadística'}, inplace=True)  # Cambiar índices
    axs[i, 2].axis('off')
    axs[i, 2].table(cellText=df_des_variable.values, colLabels=df_des_variable.columns, loc='center')
    axs[i, 2].set_title(f'Resumen Estadístico de {variable}', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
# Se comprueba, mediante test de normalidad Shapiro, si el p-valor es menor que 
# el nivel de significancia (0.05).

columnas_shapiro = ['ABV',
       'Min IBU', 'Max IBU', 'Astringency', 'Body', 'Alcohol', 'Bitter',
       'Sweet', 'Sour', 'Salty', 'Fruits', 'Hoppy', 'Spices', 'Malty',
       'review_aroma', 'review_appearance', 'review_palate', 'review_taste',
       'review_overall', 'number_of_reviews']

for column in columnas_shapiro:
    statistic, p_value = stats.shapiro(beer[column])
    
    if p_value < 0.05:
        print(f"Variable '{column}' no sigue una distribución normal.")
    else:
        print(f"Variable '{column}' sigue una distribución normal.")
```
